# K-Means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers=4, random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
X.shape, y.shape
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
km.fit(X)
```

```python
y_pred = km.predict(X)
y_pred
```

```python
# Clustering by taking center point and measuring distance for each cluster
km.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c = "red", s = 100)
```

# Noise Data

```python
X, y = make_blobs(n_samples=500, centers=4, cluster_std=2, random_state=40)
plt.scatter(X[:,0], X[:,1], c = y)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4)
km.fit(X)

y_pred = km.predict(X)

plt.scatter(X[:,0], X[:,1], c = y_pred)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c = "red", s=100)
```

## Determine number of clusters by Elbow method
### Using Inertia and distortion value

```python
error = []

for k in range(1,20):
  km = KMeans(n_clusters=k)
  km.fit(X)
  error.append(km.inertia_)
```

```python
plt.plot(range(1,20), error)
```

# Fail Cases for KMeans

## Case 1

```python
from sklearn.datasets import make_moons

X , y = make_moons(n_samples=500, noise = 0.15, random_state = 40)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
km = KMeans(n_clusters=2)
km.fit(X)

y_pred = km.predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c = "red", s = 100)
```
